```python
import pandas as pd
import matplotlib.pyplot as plt
import random
import numpy as np
from Kmeans import Kmeans
```

```python
kmeans = Kmeans()
kmeans.run()
```

```python
gbIdx = kmeans.classifiedDF.groupby('closestMean')

def findCentroid(df):
    return df.sum()/df.shape[0]

for i in range(15):
    df = kmeans.df.iloc[gbIdx.get_group(i).index]
    break
findCentroid(df)
```

```python
np.array(findCentroid(df))
```
